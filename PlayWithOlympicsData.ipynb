```python
#Data Link
#https://www.kaggle.com/datasets/heesoo37/120-years-of-olympic-history-athletes-and-results?resource=download
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load datasets
athletes = pd.read_csv('/content/drive/MyDrive/CSV/athlete_events.csv')
regions=pd.read_csv('/content/drive/MyDrive/CSV/noc_regions.csv')
```

```python
athletes.head()
```

```python
regions.head()
```

```python
#join the dataframes
#left join athletes & regions
athletes_df=athletes.merge(regions, how='left', on='NOC')
```

```python
athletes_df.head()
```

```python
#total number of riw & column 
athletes_df.shape
```

```python
#rename the region & notes column
athletes_df.rename(columns={'region':'Region', 'notes':'Notes'}, inplace=True)
```

```python
#top countries participating
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
#plot for the top 10 counteries
plt.figure(figsize=(12,6))
#plt.xticks(rotation=20)
plt.title('Overall Participating by Country')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette ='Set1')
```

```python
#Age Distribution of the participants
plt.figure(figsize=(12,6))
plt.title('Each distrbution of the athletes')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color='orange', edgecolor='white')
```

```python
#winter olympics sports
winter_sports = athletes_df[athletes_df.Season=='Winter'].Sport.unique()
winter_sports
```

```python
#summer olympics sports
summer_sports = athletes_df[athletes_df.Season=='Summer'].Sport.unique()
summer_sports
```

```python
#Male & Female Participants
gender_counts = athletes_df.Sex.value_counts()
gender_counts 
```

```python
#pie chart for male & female athletes 
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, shadow=True)
```

```python
#Total Medals
medals_count=athletes_df.Medal.value_counts()
medals_count
```

```python
#plt pie chart for medals
plt.figure(figsize=(12,6))
plt.title('Medals Distribution')
plt.pie(medals_count, labels=medals_count.index, autopct='%1.1f%%', startangle=90,shadow=True )
```

```python
#Total Female athletes who participate in each olympics
female_participants=athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
#Total Female athletes who participate in each olympics details view
womanOlympics=athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')]
womanOlympics.head(5)
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=womanOlympics, palette="Spectral")
plt.title('Woman Participants')
```

```python
#Female athletes over time
part=womanOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes over Time')
```

```python
# Gold Medal athletes
goldMedals = athletes_df[(athletes_df.Medal=='Gold')]
goldMedals.head(5)
```

```python
#take only the values are diffrent from NaN
goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
#Gold Beyond 60
goldMedals['ID'][goldMedals['Age']>60].count()
```

```python
#In Which sport come the goldmedals beyond 60 Age 
sporting_event=goldMedals['Sport'][goldMedals['Age']>60]
sporting_event
```

```python
#plot for sporting_event
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medals for Atheletes over 60 Year')
```

```python
#Gold Medals for each country
goldMedals.Region.value_counts().reset_index(name='Medal').head()
```

```python
totalGoldMedals = goldMedals.Region.value_counts().reset_index(name='Medal').head(10)
g=sns.catplot(x='index', y="Medal", data=totalGoldMedals, height=10, kind='bar', palette='rocket')
g.despine(left=True)
g.set_xlabels("Top 10 Countries")
g.set_ylabels('Number of Medals')
plt.title('Gold Medals per country')
```

```python
#Rio olympics
max_year = athletes_df.Year.max()
print(max_year)

team_names=athletes_df[(athletes_df.Year==max_year) & (athletes_df.Medal=='Gold')].Team
team_names.value_counts().head(10)
```

```python
#CountryWise Medels for the year 2016
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel('CountryWise Medals for the year 2016')
```

```python
#filter the athletes who have won a medals
not_null_medals=athletes_df[(athletes_df['Height'].notnull()) & (athletes_df['Weight'].notnull())]
```
